### FF - Ecrire un algorithme pandas python sous forme de fonction qui prends en entrée deux séries pandas et qui renvoie True si les deux séries ont au moins un élément en commun et False si non.

```python
serie1 = pd.Series(['Python', 'Java', 'Javascript', 'Kotlin'] )
serie2 = pd.Series(['Django', 'Spring', 'ReactJS', 'Kotlin'] )
serie3 = pd.Series(['Django', 'Spring', 'ReactJS', 'VueJS'] )

def common_elements(serie1, serie2):
    for i in range(len(serie1)):
        for j in range(len(serie2)):
            if serie1[i] == serie2[j]:
                return True

    return False

print(common_elements(serie1, serie2))
print(common_elements(serie1, serie3))
```
