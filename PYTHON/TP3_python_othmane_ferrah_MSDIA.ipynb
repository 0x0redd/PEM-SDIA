### c - Ecrire une fonction en python qui prends en argument une matrice numpy carrée du type nxn et  qui renvoie sa trace. On rappelle que la trace d'une matrice carrée A = (aij) i , j est le nombre  Tr(A) = a11 + a22 + ... + ann

```python
def trace_matrice(A):
    if A.shape[0] != A.shape[1]:
        raise ValueError("La matrice doit être carrée !")

    trace = 0
    for i in range(A.shape[0]):
        trace += A[i, i]
    return trace
```

```python
n = int(input("Donner la dimension n de la matrice carrée : "))

print(f"\nEntrez les {n*n} valeurs de la matrice (séparées par des espaces) :")
values = input("> ").split()


if len(values) != n*n:
    raise ValueError(f"Vous devez entrer exactement {n*n} valeurs !")


values = [float(v) for v in values]


A = np.array(values).reshape(n, n)

print("\nMatrice entrée :")
print(A)

# Trace
print("\nLa trace de la matrice est :", trace_matrice(A))
```

### e - Ecrire un programme python numpy qui permet d'empiler 2 marices numpy horizontalement, c'est-à-dire 2 tableaux ayant la même 1ère dimension (même nombre de lignes)

```python
def empiler(A , B):
    if A.shape[0] != B.shape[0]:
        raise ValueError("Les deux matrices doivent avoir le même nombre de lignes !")
    return np.hstack((A, B))
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10]])

C = empiler(A, B)

print(C)
```

### f - Ecrire un programme python numpy permettant d'empilez 2 matrices numpy verticalement, c'est-à-dire 2 tableaux ayant la même dernière dimension (même nombre de colonnes)

```python
def empiler_vertical(A , B):
    if A.shape[0] != B.shape[0]:
        raise ValueError("les deux matrice doivent avoire le meme nombre de lignes !et de colone        ")

    return np.hstack((A, B))
```

```python
D = np.array([[1, 2, 3],
              [4, 5, 6]])

E = np.array([[7, 8],
              [9, 10]])

F = empiler_vertical(D, E)

print(F)
```

### h - Ecrire un programme python-numpy permettant de remplacer les éléments diagonaux d'une  matrice numpy par des 'zéros' 

```python
import  numpy as np
def zero_diagonale(A):
    M = A.copy()
    np.fill_diagonal(M, 0)
    return M
```

```python
A = np.array([[4, 1, 2],
              [3, 5, 7],
              [8, 9, 6]])

B = zero_diagonale(A)

print("Matrice originale :")
print(A)

print("\nMatrice avec diagonale à zéro :")
print(B)
```

### 2 - Ecrire une fonction matb qui reçoit en argument un entier non nul n et renvoie un tableau numpy correspondant à la matrice ci-dessus.

```python
def matb(n):
    M = np.eye(n, dtype=int)               # diagonale principale
    M += np.eye(n, k=1, dtype=int)         # diagonale supérieure
    M += np.eye(n, k=-1, dtype=int)        # diagonale inférieure
    return M

print(matb(5))
```

### 3 - Ecrire une fonction matc qui reçoit en argument un entier non nul n et renvoie un tableau numpy correspondant à la matrice ci-dessus.

```python
def matc(n):
    # indices i et j
    i = np.arange(n).reshape(n, 1)   # colonne (n,1)
    j = np.arange(n).reshape(1, n)   # ligne  (1,n)
    M = j - i + 1                    
    M[M < 1] = 1                      # remplacer les valeurs <1 par 1
    return M.astype(int)

print(matc(6))
```
