```python
a = 0
b = 10
print(f"incrémentant la valeur de a tant qu’elle reste inférieure à celle de b")
while a < b:
    print(f"a = {a}")
    a += 1
print(f"Valeur finale : a = {a}, b = {b}")

b = 10
print(f"décrémentant la valeur de b et affichant sa valeur si elle est impaire (b non nul)")
while b != 0:
    b -= 1
    if b % 2 != 0:
        print(f"b impair = {b}")

print(f"Valeur finale : b = {b}")
```

#Exercice 11

```python
def table(base, debut, fin, inc):
    print(f"Table de {base} : de {debut} à {fin} (pas de {inc})")
    for i in range(debut, fin + 1, inc):
        print(f"{base} x {i} = {base * i}")

table(5, 1, 10, 1)
table(3, 2, 8, 2)
```

#Exercice 12

```python
def cube(x):
    return x ** 3

def volumeSphere(r):
    return (4 / 3) * math.pi * cube(r)
#test
rayon = float(input("Entrez le rayon de la sphere : "))
volume = volumeSphere(rayon)
print(f"Le volume de la sphere de rayon {rayon} est : {volume:.2f}")
```

#Exercice 13

```python
def maFonction(x):
    return 2 * x**3 + x - 5

def tabuler(fonction, borneInf, borneSup, nbPas):
    if borneInf >= borneSup:
        print("Erreur : borneInf doit être inférieure à borneSup.")
        return

    x = borneInf
    print(f"Tabulation de la fonction entre {borneInf} et {borneSup} avec un pas de {nbPas}:")

    while x <= borneSup:
        print(f"Pour x={x:.2f},f(x)={fonction(x):.4f}")
        x += nbPas


tabuler(maFonction, -3, 2, 0.5)
```

#Exercice 14

```python
def volMasseEllipsoide(a=1, b=1, c=1, masseVol=1):
    volume = (4/3) * math.pi * a * b * c
    masse = masseVol * volume
    return (volume, masse)

#test
res1=volMasseEllipsoide() # valeurs par défaut
print(f"V={res1[0]:.2f}, M={res1[1]:.2f}")
res2=volMasseEllipsoide(2, 1, 1) # 3 paramètres
print(f"V={res2[0]:.2f}, M={res2[1]:.2f}")
res3=volMasseEllipsoide(2, 3, 4, 2.5) # 4 paramètres
print(f"V={res3[0]:.2f}, M={res3[1]:.2f}")
```
